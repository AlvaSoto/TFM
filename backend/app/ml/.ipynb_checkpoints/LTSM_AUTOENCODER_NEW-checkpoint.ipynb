```python
from __future__ import annotations
import json
import os
import shutil
from dataclasses import dataclass
from datetime import datetime
from pathlib import Path
from typing import Dict, List, Sequence, Tuple
import joblib
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_fscore_support, roc_auc_score
from sklearn.preprocessing import StandardScaler
from tensorflow.keras import callbacks, layers, models, optimizers

#-- Class to avoid Json errors --#
class CustomJSONEncoder(json.JSONEncoder):
    def default(self, o):
        if isinstance(o, Path):
            return str(o)
        if isinstance(o, np.ndarray):
            return o.tolist()
        if isinstance(o, np.number):
            return o.item()
        return super().default(o)


@dataclass
class PipelineConfig:
    data_path: Path

    #Key parameters
    sequence_length: int =96 #24h (96 intervals of 15 mins)
    sequence_stride: int=2 #Moving the window every 30 mins

    train_split_ratio: float=0.9 #I use the 90% of the time to train, 10% to test
    validation_split: float=0.15 #From the 90%, i book 15% for partial examns (validation)

    latent_dim: int=64 #neck bottle is wider (64 neurons) 
    lstm_units: Sequence[int] = (256, 128, 64) #Making the model smarter

    batch_size: int=256
    max_epochs: int=150
    patience: int=15 #if it does not improve in 15 loops it stops
    threshold_percentile: float=99.0

    #Lifevest route
    base_output_dir: Path = Path.home() / "TFM_Project_Runs"

config = PipelineConfig(data_path=Path('../../data/mixed_population_dataset_160_households.csv'))
tf.random.set_seed(42)
np.random.seed(42)

print("✅ Block 1 completed. Config ready.")
print(f"📂 The results will be stored in: {config.base_output_dir}")
    
```

```python
#Block 2 Preparing the backup folders and the functions

def prepare_artifacts_dirs(config: PipelineConfig) -> Dict[str, Path]:
    #Using a fixed folder to be able to resume always
    project_dir = config.base_output_dir / "Run_Ultimate"

    paths = {
        "project_dir": project_dir,
        "checkpoint_dir": project_dir / "checkpoints",
        "backup_dir": project_dir / "backup_state",
        "results_dir": project_dir / "results",
        "scaler_path": project_dir / "scaler.joblib",
        "history_path": project_dir / "training_log.csv"
    }

    for p in [paths["project_dir"], paths["checkpoint_dir"], paths["backup_dir"], paths["results_dir"]]:
        p.mkdir(parents=True, exist_ok=True)

    return paths

#Returning tuples to avoid changes, also faster than lists
def engineer_features(df: pd.DataFrame) -> Tuple[pd.DataFrame, List[str]]:
    print("Starting feature engineering")
    df = df.copy()
    df["timestamp"] = pd.to_datetime(df["timestamp"])
    df = df.sort_values(["household_id", "timestamp"]).reset_index(drop=True)

    for col, period in [('hour', 24), ('dayofweek', 7)]:
        df[f'{col}_sin'] = np.sin(2 * np.pi * getattr(df['timestamp'].dt, col) / period)
        df[f'{col}_cos'] = np.cos(2 * np.pi * getattr(df['timestamp'].dt, col) / period)

    df['is_weekend'] = (df['timestamp'].dt.dayofweek >= 5).astype(int)
    # is_night: periodo donde suele haber fugas silenciosas (00:00 - 05:00)
    df['is_night'] = df['timestamp'].dt.hour.between(0, 5, inclusive="both").astype(int)

    grouped = df.groupby("household_id", sort=False)['consumption_l']

    # Ventanas Deslizantes (Captura tendencias y variabilidad)
    for window in [4, 8, 12, 24, 48, 96]:
        rolling = grouped.rolling(window=window, min_periods=1)
        rmean = rolling.mean().reset_index(level=0, drop=True)
        rstd = rolling.std().reset_index(level=0, drop=True)
        
        df[f"rolling_mean_{window}"] = rmean
        df[f"rolling_std_{window}"] = rstd
        df[f"rolling_cv_{window}"] = rstd / (rmean + 1e-6) # Coeficiente de variación
        df[f'diff_from_rolling_mean_{window}'] = df['consumption_l'] - rmean
        
    df.fillna(0, inplace=True)

    # Lags (Comparación con ayer y la semana pasada)
    for lag in [96, 96*7]:
        df[f'consumption_lag_{lag}'] = grouped.shift(lag).fillna(0)
    
    feature_cols = [col for col in df.columns if col not in ['timestamp', 'is_leak', 'household_id']]
    return df, feature_cols


def create_sequences(df, feature_cols, seq_length, stride):
    sequences, labels, metadata = [], [], []
    
    print(f"Procesando {df['household_id'].nunique()} hogares...")
    for household_id, group in df.groupby("household_id"):
        group = group.sort_values("timestamp")
        values = group[feature_cols].values
        leak_flags = group["is_leak"].values
        timestamps = group["timestamp"].values

        if len(group) < seq_length: continue
        
        # Optimización vectorial para velocidad
        num_seq = (len(group) - seq_length) // stride + 1
        starts = np.arange(num_seq) * stride
        ends = starts + seq_length
        
        for s, e in zip(starts, ends):
            sequences.append(values[s:e])
            # Si hay al menos un 1 en la ventana, se marca como fuga (aunque entrenaremos solo con 0)
            labels.append(int(np.max(leak_flags[s:e])))
            metadata.append({
                "household_id": household_id,
                "end_timestamp": timestamps[e-1]
            })

    return np.stack(sequences), np.array(labels, dtype=int), metadata

print("✅ Block 2 completed. Functions ready.")

            
```

```python
#Block 3: Load data an create secuences (Memory lets see if it can handle it)
#preapring the folders
paths = prepare_artifacts_dirs(config)
#Load the csv
print(f"Reading the dataset from {config.data_path}")
df = pd.read_csv(config.data_path)
df, feature_cols = engineer_features(df)

print("Scaling the data...")
scaler = StandardScaler()
df[feature_cols] = scaler.fit_transform(df[feature_cols])
df[feature_cols] = df[feature_cols].astype(np.float32)#Reduce the RAM weight

joblib.dump(scaler, paths["scaler_path"])
print(f"Scaler saved")

print(f"Generating sequences with Stride={config.sequence_stride}...")
sequences, labels, metadata = create_sequences(df, feature_cols, config.sequence_length, config.sequence_stride)

print(f"✅ Block 3 completed.")
print(f"📊 Sequences generated: {len(sequences):,}")
print(f"   Data format: {sequences.shape}")

```

```python
#Block 4: Data Division
end_times = np.array([meta["end_timestamp"] for meta in metadata])
sorted_idx = np.argsort(end_times)

#Punto de corte, 90% to train/validate and 10% to test
split_pt = int(len(end_times) * config.train_split_ratio)
train_mask = np.zeros(len(end_times), dtype=bool)
train_mask[sorted_idx[:split_pt]] = True

X_train_full = sequences[train_mask]
y_train_full = labels[train_mask]
X_test = sequences[~train_mask]
y_test = labels[~train_mask]

#Key filter: The autoencoder just sees the Normal data (label 0) to learn what is normal
X_train_normal = X_train_full[y_train_full == 0]

#Split the validation
val_size = int(len(X_train_normal) * config.validation_split)
X_val = X_train_normal[-val_size:]
X_train = X_train_normal[:-val_size]

print(f"✅ Block 4 completed.")
print(f"   Training (Only normal): {len(X_train):,} Sequences")
print(f"   Validation (Only normal)):    {len(X_val):,} Sequences")
print(f"   Test (Mix Normal/leak):   {len(X_test):,} Sequences")
```

```python
#Block 5
def build_lstm_autoencoder(input_shape, latent_dim, lstm_units):
    inputs = layers.Input(shape=input_shape)
    x = inputs
    
    # --- ENCODER ---
    for units in lstm_units:
        x = layers.LSTM(units, return_sequences=True)(x)
        x = layers.BatchNormalization()(x) # <--- Mejora: Estabilidad
        x = layers.Dropout(0.2)(x)
        
    x = layers.LSTM(lstm_units[-1] // 2, return_sequences=False)(x)
    x = layers.BatchNormalization()(x)
    
    # --- CUELLO DE BOTELLA (LATENT SPACE) ---
    latent = layers.Dense(latent_dim, activation="relu", name="bottleneck")(x)

    # --- DECODER ---
    x = layers.RepeatVector(input_shape[0])(latent)
    for units in reversed(lstm_units):
        x = layers.LSTM(units, return_sequences=True)(x)
        x = layers.BatchNormalization()(x)
        x = layers.Dropout(0.2)(x)
        
    outputs = layers.TimeDistributed(layers.Dense(input_shape[1]))(x)

    model = models.Model(inputs, outputs, name="lstm_autoencoder")
    # LR reducido para precisión fina con datos complejos
    model.compile(optimizer=optimizers.Adam(1e-4), loss="mse") 
    return model

model = build_lstm_autoencoder(
    (config.sequence_length, len(feature_cols)),
    config.latent_dim,
    config.lstm_units
)
model.summary()
print("✅ Block 5 completed. Model compiled.")
```

```python
# %% BLOQUE 6: ENTRENAMIENTO CON BACKUP
# Callbacks para seguridad y optimización
backup_cb = callbacks.BackupAndRestore(backup_dir=paths["backup_dir"])
checkpoint_cb = callbacks.ModelCheckpoint(
    filepath=paths["checkpoint_dir"] / "best_model.keras",
    monitor="val_loss", save_best_only=True, verbose=1
)
csv_log = callbacks.CSVLogger(paths["history_path"], append=True)
es = callbacks.EarlyStopping(monitor="val_loss", patience=config.patience, restore_best_weights=True)
rlr = callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=5, verbose=1)

print("🚀 Iniciando entrenamiento...")
print("ℹ️ Puedes detener este bloque cuando quieras. Al re-ejecutarlo, continuará.")

history = model.fit(
    X_train, X_train,
    validation_data=(X_val, X_val),
    epochs=config.max_epochs,
    batch_size=config.batch_size,
    callbacks=[backup_cb, checkpoint_cb, csv_log, es, rlr],
    shuffle=True,
    verbose=2
)
print("✅ Bloque 6 completado. Entrenamiento finalizado.")
```

```python
# %% BLOQUE 7: EVALUACIÓN Y RESULTADOS
def compute_reconstruction_errors(model, X):
    # Batch size grande para predecir rápido
    return np.mean(np.square(X - model.predict(X, batch_size=512, verbose=0)), axis=(1, 2))

# 1. Cargar MEJOR modelo (no el último de la RAM)
print("Cargando el mejor modelo guardado...")
best_model = models.load_model(paths["checkpoint_dir"] / "best_model.keras")

# 2. Calcular Umbral
print("Calculando errores de validación...")
val_errors = compute_reconstruction_errors(best_model, X_val)
threshold = float(np.percentile(val_errors, config.threshold_percentile))
print(f"🎯 Umbral detectado (p{config.threshold_percentile}): {threshold:.6f}")

# 3. Evaluar en Test
print("Evaluando en Test...")
test_errors = compute_reconstruction_errors(best_model, X_test)
y_pred = (test_errors >= threshold).astype(int)

# 4. Métricas
metrics = {
    "precision": float(precision_recall_fscore_support(y_test, y_pred, average="binary")[0]),
    "recall": float(precision_recall_fscore_support(y_test, y_pred, average="binary")[1]),
    "f1": float(precision_recall_fscore_support(y_test, y_pred, average="binary")[2]),
    "roc_auc": float(roc_auc_score(y_test, test_errors)),
    "confusion_matrix": confusion_matrix(y_test, y_pred).tolist(),
    "classification_report": classification_report(y_test, y_pred, output_dict=True)
}

# 5. Guardar
final_results = {
    "config": vars(config),
    "threshold": threshold,
    "metrics": metrics,
    "paths": {k: str(v) for k,v in paths.items()}
}

res_file = paths["results_dir"] / f"final_metrics_{datetime.now().strftime('%Y%m%d_%H%M')}.json"
res_file.write_text(json.dumps(final_results, indent=2, cls=CustomJSONEncoder))

print("="*50)
print(f"📊 RESULTADOS:")
print(f"   ROC-AUC: {metrics['roc_auc']:.4f}")
print(f"   Recall:  {metrics['recall']:.4f}")
print(f"   Precisión: {metrics['precision']:.4f}")
print(f"💾 Informe guardado en: {res_file}")
print("="*50)
```
